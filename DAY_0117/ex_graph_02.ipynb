(1) 모듈 로딩 <hr>

```python
import matplotlib.pyplot as plt        # 그래프 관련 모듈
import matplotlib as mpl               # 패키지 관련
import matplotlib.font_manager as fm   # 패키지 리소스 즉 자원 중 폰트 관련 모델

import pandas as pd
```

(2) 데이터 준비 <hr>

```python
# 2D 그래프를 그리기 위한 데이터 준비
x = list(range(1, 100, 10))
y = [d**2 for d in x]
```

(3) 그래프 그리기 <hr>

- [기본] 라인 / 선 그래프
    * 시간의 변화를 표현해주는 그래프
    * 연속적 / 순서 있는 데이터를 시각화할 경우 주로 사용됨

```python
# plot(data) => y축 데이터로 사용
# plot(data1, data2) => x축에 data1, y축에 data2
# 반환값 : 라인 / 선 그래프 객체 리스트에 담겨서 반환됨
ret = plt.plot(x, y, 'r-o')
type(ret), len(ret), type(ret[0])

# 파이썬에서 표준출력 기능 내장 함수 => print()와 동일
plt.show()
```

```python
ret[0].get_xdata()
```

- 라인 / 선 그래프에서 형식 즉 포맷지정 => str 타입의 포맷 설정
    * '색상형태선스타일' ==> 'ro--' : red  circle  dot-line
    * 대표적으로 많이 사용되는 색상 : r - red, g - green, b - blue, k - black
    * 대표적으로 많이 사용되는 형태 : o - 동그라미, ^ - 위로 삼각형, < - 왼쪽 삼각형, > - 오른쪽 삼각형 v - 아래로 삼각형
    * 대표적으로 많이 사용되는 선스타일 : - 실선, -- 점선, -.-.

```python
ret = plt.plot(x, y, 'g<--')
```

```python
x2 = [d // 2 for d in x]
y2 = [d // 2 for d in y]
```

### X,Y축 레이블 설정 및 눈금 설정
- x, y축 데이터에대한 설명 => plt.xlabel(), plt.ylabel()
- 눈금 간격, 출력 형태 설정 => plt.xticks(), plt.yticks(), plt.tick_param()

```python
# 그래프 그리기
plotList = plt.plot(y, 'yo--', y2, 'g^')
plt.title('[테스트]')

# x, y축 레이블
plt.xlabel('값', loc = 'right', fontsize = 'xx-large', color = 'red')
plt.ylabel('결과', loc = 'top', fontsize = 'xx-large', color = 'blue', rotation = 'horizontal')

# x, y 눈금 설정
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E',], color = 'red')
plt.yticks(color = 'blue')
```

### 범례 설정
- 여러개의 그래프를 표시하는 경우 그래프별 식별을 위해서 사용

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'g^')
plt.title('[테스트]')

# x, y축 레이블
plt.xlabel('값', loc = 'right', fontsize = 'xx-large', color = 'red')
plt.ylabel('결과', loc = 'top', fontsize = 'xx-large', color = 'blue', rotation = 'horizontal')

# x, y 눈금 설정
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E',], color = 'red')
plt.yticks(color = 'blue')

# 범례 설정
plt.legend(['red 결과', 'green 결과'], loc = 10)

# 그래프 안에 텍스트 출력
plt.text(2, 2000, 'Happy New Year')
```

### 텍스트 추가
- 그래프 내에 위치 지정 후 텍스트 추가 설정 => plt.text(시작x, 시작y, 출력테스트)

```python
# 그래프 그리기
plotList = plt.plot(y, 'ro--', y2, 'b^')
plt.title('[테스트]')

# x, y축 레이블
plt.xlabel('값', loc = 'right', fontsize = 'xx-large', color = 'red')
plt.ylabel('결과', loc = 'top', fontsize = 'xx-large', color = 'blue', rotation = 'horizontal')

# x, y 눈금 설정
plt.xticks(ticks = [0, 2, 4, 6, 8], labels = ['A', 'B', 'C', 'D', 'E',], color = 'red')
plt.yticks(color = 'blue')

# 범례 설정
plt.legend(['red 결과', 'green 결과'])

# 그래프 안에 텍스트 출력
plt.text(2, 2000, 'Happy New Year', backgroundcolor = 'yellow')

box = {'facecolor' : '.75', 'edgecolor' : 'k', 'boxstyle' : 'round', 'alpha' : 0.2}
plt.text(2, 4000, 'Happy New Year', bbox = box)
```

```python
# 화살표 추가

X = np.linspace(-4, 4, 1024)
Y = .25 * (X + 4.) * (X + 1.) * (X - 2.)

plt.annotate('Brackmard minimun',
             ha = 'center', va = 'bottom',
             xytext = (-1.5, 3.),
             xy = (0.75, -2.7),
             arrowprops = {'facecolor' : 'black', 'shrink' : 0.05})

plt.plot(X, Y)
plt.show()
```

```python
# 범례 추가
X = np.linspace(0, 6, 1024)
Y1 = np.sin(X)
Y2 = np.cos(X)

plt.xlabel('X')
plt.ylabel('Y')

plt.plot(X, Y1, c = 'k', lw = 3., label = 'sin(X)')
plt.plot(X, Y2, c = '.5', lw = 3., ls = '--', label = 'cos(X)')

plt.legend()
plt.show()
```

```python
# 격자 추가
X = np.linspace(-4, 4, 1024)
Y = .25 *(X + 4.) * (X + 1.) * (X - 2.)

plt.plot(X, Y, c = 'k')
plt.grid(True, lw = 2, ls = '--', c = '.75')
plt.show()
```
